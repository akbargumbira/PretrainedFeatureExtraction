```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = [5, 4]
plt.rcParams["font.size"] = 12


sigmoid = lambda x: 1 / (1 + np.exp(-x))
dsigmoid = lambda x: sigmoid(x) * (1-sigmoid(x))
tanh = lambda x: np.tanh(x)
dtanh = lambda x: (1-tanh(x)**2)
relu = lambda x: np.maximum(0, x)
drelu = lambda x: np.where(x > 0, 1, 0) 
leaky_relu = lambda x: np.maximum(0.1*x, x)
dleaky_relu = lambda x: np.where(x > 0, 1, 0.1) 
y=np.linspace(-5,5,1000)

def plot_activation_function(x, func, dfunc, title):    
    plt.plot(y, func(y),'r', linewidth=2)
    plt.plot(y, dfunc(y),'b--', linewidth=2)
    
    plt.axhline(0, color='#808080')
    plt.axvline(0, color='#808080')
    plt.grid()
    plt.xlabel(r'x')
    plt.ylabel(r'f (x)')
    plt.title(title)
    plt.legend()
    plt.show()
    
plot_activation_function(y, sigmoid, dsigmoid, 'Sigmoid')
plot_activation_function(y, tanh, dtanh, 'tanh')
plot_activation_function(y, relu, drelu, 'ReLU')
plot_activation_function(y, leaky_relu, dleaky_relu, r'Leaky ReLU ($\alpha$ = 0.1)')
```

```python
k=2
alpha=10**-4
beta=0.75

%matplotlib inline
plt.rcParams["figure.figsize"] = [8, 4]
plt.rcParams["font.size"] = 14

f = lambda x: 1 / (k + alpha * x)**beta
y=np.linspace(0,1000000,1000)

plt.plot(y, f(y),'r', linewidth=2)
plt.axhline(0, color='#808080')
plt.axvline(0, color='#808080')
plt.grid()
plt.xlabel(r'$x = \sum_{}^{}(a^{j}_{x,y})^2$', fontsize=16)
h = plt.ylabel(r'$y= \frac{1}{(k + \alpha x)^ \beta}$', fontsize=22)
plt.legend()
plt.show()
```
